```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentences = [
    'I absolutely love this product; it works perfectly.',
    'I absolutely love wonderful and productive day.',
    'The movie was fantastic and full of excitement.',
    'I am feeling very happy and grateful right now.',
    'This is the best decision I have ever made.'
]
```

```python
sentences
```

```python
vocabulary_size = 10000
```

```python
one_hot_representation = [one_hot(words, vocabulary_size)for words in sentences]
```

```python
one_hot_representation
```

```python
## word embeding representation
from tensorflow.keras.layers import Embedding, Input
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
## make all sentences same size
sent_size = 10
embed_docs = pad_sequences(one_hot_representation, padding='post', maxlen=sent_size)
```

```python
embed_docs
```

```python
dimensions = 10
model = Sequential()
model.add(Input(shape=(sent_size,)))
model.add(Embedding(
    input_dim=vocabulary_size,
    output_dim=dimensions
))
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.predict(embed_docs)
```
